# Matplot Library

* Matplotlib is a library used for creating visualization
* Matplotlib is a python 2D plotting
* It is a Data Science library
* Types of plotting
     * Line
     * Bar Chart
     * Area Plot
     * Scatter plot
     * Pie Chart
     * Histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Line plotting
    * plt.plot()

```python
plt.plot([1,2,3,4],[1,6,7,8])   # (1,1)(2,6)(3,7)(4,8)
plt.show()
```

```python
plt.plot([3,4,5,6],[5,7,8,9],color='red')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plotting')
```

```python
x = np.linspace(0,2,10)
x
```

```python
plt.plot(x,x,label='Line',color='brown')
plt.plot(x,x**2,label='Square',color='blue')
plt.plot(x,x**3,label='Cube',color='r')
plt.legend()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plotting')
```

```python
x1 = np.random.randint(0,10,5)
print(x1)
y1 = np.random.randint(10,50,5)
print(y1)
```

# SCatter Plot
   * plt.scatter()

```python
plt.scatter([1,2,3,4],[5,6,7,8],color='r')
```

```python
plt.scatter([1,2,3,4,5,6,7,8,9],[4,5,8,1,2,3,7,8,9],color='g')
```

```python
plt.scatter(x1,y1,color='blue')
```

# Bar Chart 
    * plt.bar()

```python
plt.bar([1,2,3,4],[5,6,7,8],width=0.5)
```

```python
x2 = np.arange(4)
print(x2)
plt.bar(x2,[78,90,56,78],width=0.3,color='red')
plt.bar(x2+0.3,[67,8,79,90],width=0.3,color='pink')
```

# Histogram
   - Histogram is used to show the frequency Distributon
   - plt.hist()

```python
a = [2,3,4,5,6,7,8,9,1,2,3,7,8,9,3]
bins = 5
plt.hist(a,bins,width=0.5)
# bins()  - represents intervals (or) range of values on x-axis
```

# Area Plot
  * plt.stackplot()

```python
days = [1,2,3,4]
working = [4,5,6,7]
sleeping = [7,8,1,5]
plt.stackplot(days,working,sleeping,colors=['g','r'])
plt.show()
```

# Pie Chart
  * plt.pie()

```python
rating = [5,4.8,8,9,6]
names = ['python','c','java','c++','oracle']
plt.pie(rating,labels=names)
plt.legend()
plt.show()
```
